<h2>Essential data cleaning tasks:</h2>

Fill Missing Values: Replaces missing values in the total_clean_air_amount column with its mean.

Convert Data Types: Converts the programme_adoption_date column to a datetime format for easier date manipulation.

Remove Duplicates: Eliminates duplicate rows from the dataset to ensure data integrity.

```python
# Example: Fill missing values in a column with the mean
df['total_clean_air_amount'] = df['total_clean_air_amount'].fillna(df['total_clean_air_amount'].mean())


# Example: Convert date columns to datetime
df['programme_adoption_date'] = pd.to_datetime(df['programme_adoption_date'])

# Remove duplicates
df.drop_duplicates(inplace=True)
```

<h2>Data aggregation and ranking:</h2>

Group and Summarize: Groups data by ms_name (country names) and calculates the total total_clean_air_amount for each country.
                                                                                                           
Sort Results: Sorts the aggregated data in descending order based on total_clean_air_amount to identify countries with the highest totals.
                                                                                                           
Preview Data: Displays the top results for quick analysis.

```python
country_clean_air = df.groupby('ms_name')['total_clean_air_amount'].sum().reset_index()

# Sort the results in descending order
country_clean_air = country_clean_air.sort_values(by='total_clean_air_amount', ascending=False)

country_clean_air.head()
```

<p>Romania has the highest total, while Austria and Belgium have no clean air contributions recorded, as indicated by zeros. The data provides a clear ranking of countries based on their clean air amounts.</p>

<h2>Vizualisation:</h2>
A bar plot to display the total clean air amount for each country, allowing for a comparison of clean air contributions across countries.
A pie chart to visualize the distribution of different program statuses, showing the proportion of each status in the dataset.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='total_clean_air_amount', y='ms_name', data=country_clean_air)
plt.title('Total Clean Air Amount by Country')
plt.xlabel('Total Clean Air Amount')
plt.ylabel('Country')
plt.show()

status_counts = df['programme_status'].value_counts()

plt.figure(figsize=(8, 8))
status_counts.plot.pie(autopct='%1.1f%%', startangle=90, cmap='Set2')
plt.title('Distribution of Programme Status')
plt.ylabel('')
plt.show()
```

The pie chart illustrates the distribution of program adoption statuses:

87.8% of the programs were adopted by the European Commission (EC), showing the EC's predominant involvement.

7.2% were adopted by Member States (MS), representing a smaller share of adoption.

4.9% were adopted by the EC with the "TF Accepted" status, indicating a specialized adoption type involving task forces or specific conditions.

<h2>Creating a time series plot showing the total clean air amount by year:</h2>

The data is first grouped by year (extracted from programme_adoption_date), and the sum of total_clean_air_amount is calculated for each year.
    
A line plot is generated to visualize how the total clean air amount has changed over time, with the x-axis representing the years and the y-axis representing the total amount.
    
The plot includes a title and labeled axes for clarity, and the figure size is set to 10x6 inches for better readability.

```python
time_series_yearly = df.groupby(df['programme_adoption_date'].dt.year)['total_clean_air_amount'].sum()
time_series_yearly.plot(kind='line', title="Funding Over Time (Yearly)", xlabel="Year", ylabel="Total Clean Air Amount", figsize=(10, 6))
```

The data is grouped by the un_sdg_tracking column, and the total clean air amount is summed for each SDG category. A bar chart is created to display the total funding for each SDG, with the chart titled "Funding by SDG" and a figure size of 10x6 inches for better readability.

```python
sdg_funding = df.groupby('un_sdg_tracking')['total_clean_air_amount'].sum()
sdg_funding.plot(kind='bar', figsize=(10, 6), title="Funding by SDG")
```

The total clean air funding is aggregated by region using the category_region_name column. A pie chart is then displayed, illustrating the distribution of funding across different regions, with percentages shown and the title "Funding by Region."

```python
region_funding = df.groupby('category_region_name')['total_clean_air_amount'].sum()
region_funding.plot(kind='pie', autopct='%1.1f%%', title="Funding by Region")
```

A new column, utilization_rate, is created by dividing the spent_eucleanairamount_est by the decided_eucleanairamount_est to calculate the funding utilization rate. The data is then grouped by ms_name (country), and the average utilization rate for each country is calculated. A bar chart is generated to display the average utilization rate by country, with the title "Funding Utilization Rate by Country."

```python
df['utilization_rate'] = df['spent_eucleanairamount_est'] / df['decided_eucleanairamount_est']
utilization_summary = df.groupby('ms_name')['utilization_rate'].mean()
utilization_summary.plot(kind='bar', title="Funding Utilization Rate by Country")
```
